```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
auto = pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

```python
sns.displot(auto['citympg'])
plt.show()
```

```python
sns.displot(auto['enginesize'],kde=False,rug=True)
plt.show()
```

```python
sns.jointplot(x=auto['enginesize'], y=auto['horsepower'])
plt.show()
```

```python
#kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
sns.jointplot(x=auto['enginesize'], y=auto['horsepower'],kind ="hex")
plt.show()
```

```python
sns.jointplot(x=auto['enginesize'], y=auto['horsepower'],kind ="reg")
plt.show()
```

```python
sns.pairplot(auto[['enginesize','horsepower','citympg']])
```

```python
sns.stripplot(x=auto['fueltype'],y=auto['horsepower'],jitter= True)
```

```python
sns.swarmplot(x=auto['fueltype'],y=auto['horsepower'])
```

```python
sns.boxplot(x=auto['doornumber'],y=auto['horsepower'],hue=auto['fueltype'])
```

```python
sns.barplot(x=auto['carbody'],y=auto['horsepower'],hue=auto['enginelocation'])
```

```python
sns.countplot(x=auto['carbody'],hue=auto['enginelocation'])
```

```python
sns.pointplot(x=auto['fuelsystem'],y=auto['horsepower'],hue=auto['doornumber'])
```
